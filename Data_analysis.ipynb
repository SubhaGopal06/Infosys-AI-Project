```python
import pandas as pd

# Load the datasets
all_tickets_path = '/content/all_tickets.csv'
customer_support_tickets_path = '/content/customer_support_tickets.csv'

all_tickets = pd.read_csv(all_tickets_path)
customer_support_tickets = pd.read_csv(customer_support_tickets_path)

# Function for EDA
def perform_eda(df, dataset_name):
    print(f"\n{'='*30} EDA for {dataset_name} {'='*30}\n")

    # Display first few rows
    print("First 5 rows:\n", df.head(), "\n")

    # Display basic info
    print("Basic Info:\n")
    print(df.info(), "\n")

    # Check for missing values
    print("Missing Values:\n", df.isnull().sum(), "\n")

    # Statistical summary for numerical columns
    print("Statistical Summary for Numerical Columns:\n", df.describe(), "\n")

    # Statistical summary for categorical columns
    categorical_summary = df.select_dtypes(include=['object']).describe()
    print("Statistical Summary for Categorical Columns:\n", categorical_summary, "\n")

    # Value counts for categorical columns
    print("Value Counts for Categorical Columns:\n")
    for col in df.select_dtypes(include=['object']).columns:
        print(f"Column: {col}\n", df[col].value_counts(), "\n")

# Perform EDA on both datasets
perform_eda(all_tickets, "all_tickets.csv")
perform_eda(customer_support_tickets, "customer_support_tickets.csv")
```
